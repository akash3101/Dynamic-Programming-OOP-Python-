```python
bills = [3,5,7,9]
allSolutions = []
def buyBirds(bills,target,highest=1,sol = []):
    global allSolutions
    if sum(sol) == target:
        allSolutions.append(sol)
        return
    if sum(sol) > target:
        return
    for bill in bills:
        if bill >= highest:
            newSol = sol[:]
            newSol.append(bill)
            buyBirds(bills,target,bill,newSol)
    return
               
buyBirds(bills,100)

counter = {}
for s in range(len(allSolutions)):
    counter[len(allSolutions[s])] = allSolutions[s]

print(allSolutions)
print("------")
minAmount = counter[min(counter.keys())]
print(minAmount)

numBirds = {}
Pig = 0
Cra = 0
Gee = 0
Pea = 0

for k in minAmount:
    if k == 3:
        Pig += 5
    if k == 5:
        Cra += 7
    if k == 7:
        Gee += 9
    if k == 9:
        Pea += 3

numBirds["Pigeons"] = Pig
numBirds["Cranes"] = Cra
numBirds["Geese"] = Gee
numBirds["Peacock"] = Pea

print("----")
print(numBirds)
```
